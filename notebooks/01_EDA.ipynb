```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

print(df.shape)       # (7043, 21)
print(df.dtypes)
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df['Churn'].value_counts().plot(kind='bar', color=['green','red'])
plt.title('Distribution du Churn')
plt.show()
```

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
df[num_cols].hist(figsize=(12,4), bins=30)
plt.show()
```

```python
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn selon le type de contrat')
plt.show()
```

```python
df_num = df.copy()
df_num['Churn_bin'] = (df_num['Churn'] == 'Yes').astype(int)
df_num['TotalCharges'] = pd.to_numeric(df_num['TotalCharges'], errors='coerce')
sns.heatmap(df_num[num_cols + ['Churn_bin']].corr(), annot=True, cmap='coolwarm')
plt.show()
```
